#### 3.1 Preprocesamiento con `ColumnTransformer`

- Convierta las columnas mal leidas a sus tipos correspondientes (float, str, etc...)
- Genere un `ColumnTransformer` que:
  - Preprocese datos categóricos y ordinales.
  - Escale/estandarice datos numéricos.
  - Uitlice `.set_output(transform="pandas")` sobre su `ColumnTransformer` para setear el formato de salida a de las transformaciones a pandas.

- Luego, pruebe las transformaciones utilizando `fit_transform`.

- Posteriormente, ejecute un Holdout que le permita más adelante evaluar los modelos.



Primero arreglemos las columnas que tienen datos mal leídos/ingresados (datos infactibles). Del archivo profile.html adjunto en la tarea (el cual corresponde a un informe de EDA de los datos) notamos lo siguiente respecto a las columnas:

- age: Hay 104 clientes que tienen edades negativas (-500) y 5 clientes con edades del orden de 8600. Esto no tiene sentido, y debido a que son pocos datos (109) en comparación al total (12.500) decidimos eliminarlos.

- num_bank_accounts: Hay 4 clientes con una cantidad de cuentas bancarias negativa (-1) y 5 clientes con una cantidad de cuentas bancarias del orden de 1700. Esto no tiene sentido, y debido a que son pocos datos (9) en comparación al total (12.500) decidimos eliminarlos.

- num_credit_card: Hay 6 clientes con una cantidad de tarjetas de crédito del orden de 1400. Esto no tiene sentido, y debido a que son pocos datos (6) en comparación al total (12.500) decidimos eliminarlos.

- interest_rate: Hay 5 clientes con una tasa de interés en tarjeta de crédito del orden de 5700. Esto no tiene sentido, y debido a que son pocos datos (5) en comparación al total (12.500) decidimos eliminarlos.

- num_of_loan: Hay 500 clientes con una cantidad de préstamos otorgados por el banco negativo (-500) y 5 clientes con una cantidad de préstamos otorgados del orden de 1400. Esto no tiene sentido, y debido a que son pocos datos (505) en comparación al total (12.500) decidimos eliminarlos.

- delay_from_due_date: Hay 88 clientes con un promedio de días atrasados de pago negativo. Esto no tiene sentido, y debido a que son pocos datos (88) en comparación al total (12.500) decidimos eliminarlos.

- num_of_delayed_payment: Hay 81 clientes con un promedio de cantidad de pagos atrasados negativo y 5 clientes con un promedio de cantidad de pagos atrasados del orden de 4000. Esto no tiene sentido, y debido a que son pocos datos (86) en comparación al total (12.500) decidimos eliminarlos.

Arreglemos las columnas de acuerdo a esto y filtremos los datos adecuados:


```python
mask = (
    (df['age']>0)
    & (df['age']<8600)
    & (df['num_bank_accounts']>=0)
    & (df['num_bank_accounts']<1700)
    & (df['num_credit_card']<1400)
    & (df['interest_rate']<5700)
    & (df['num_of_loan']>=0)
    & (df['num_of_loan']<1400)
    & (df['delay_from_due_date']>=0)
    & (df['num_of_delayed_payment']>=0)
    & (df['num_of_delayed_payment']<4000)
)

df_limpio = df[mask]

print('Cantidad de datos originales: {}'.format(len(df)))
print('Cantidad de datos limpios: {}'.format(len(df_limpio)))
print('Cantidad y porcentaje de datos eliminados: {} ({}%)'
.format(len(df)-len(df_limpio), int(round(100*(len(df)-len(df_limpio))/len(df),0))))
```

Analicemos datos nulos ahora (cantidad y porcentaje de datos nulos por cada atributo):

```python
cantidad_nulos = df_limpio.isnull().sum()
print(cantidad_nulos)
```

```python
porcentaje_nulos = round(df_limpio.isnull().mean() * 100,0).astype(int)
print(porcentaje_nulos)
```

Notemos que las columnas categóricas (customer_id, occupation, payment_of_min_amount y payment_behaviour) no tienen datos nulos, pero las numéricas si tienen (en particular 6 columnas del total de 22). Dado que los porcentajes de nulos de estas variables numéricas son en general bajos (lo máximo que hay en una columna es de 15% seguido de 9%, y el resto de 4 columnas tiene un porcentaje menor o igual al 5%) y para no añadir sesgo al imputar aquellos datos nulos por la media del atributo respectivo, decidimos eliminar estas filas con datos nulos de nuestro análisis:   

```python
df_final = df_limpio.dropna()

print('Cantidad de datos originales: {}'.format(len(df_limpio)))
print('Cantidad de datos finales: {}'.format(len(df_final)))
print('Cantidad y porcentaje de datos (nulos) eliminados: {} ({}%)'
.format(len(df_limpio)-len(df_final), int(round(100*(len(df_limpio)-len(df_final))/len(df_limpio),0))))
```

Ahora creamos la variable objetivo y la matriz de características (es importante destacar que la columna customer_id se excluirá pues al ser un ID no tiene sentido incorporarla en la predicción de riesgo crediticio para un cliente general):

```python
y = df_final['credit_score']
X = df_final.drop(columns=['credit_score', 'customer_id'])
```

Veamos el balance de clases:

```python
y.value_counts()
```

Notemos que la clase positiva 1 (cliente de riesgo crediticio) es la minoritaria (tiene la mitad de ejemplos que la clase negativa 0 mayoritaria de clientes sin riesgo crediticio).

Ahora generamos el column transformer de preprocesamiento. Las 3 columnas categóricas de X (que son nominales) serán preprocesadas por OneHotEncoder, mientras que las 17 columnas restantes numéricas de X serán preprocesadas por MinMaxScaler (es decir se normalizarán los datos, lo cual está bien justificado al haber eliminado los outliers principales con el filtro de datos que realizamos):



```python
# Columnas según su tipo
categoricas = X.select_dtypes(include=['object']).columns
numericas = X.select_dtypes(include=['number']).columns

# Librerías
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.compose import ColumnTransformer

# Encoders
encoder_categoricas = OneHotEncoder(sparse=False)
encoder_numericas = MinMaxScaler()

# Column Transformer de preprocesamiento
preprocesamiento = ColumnTransformer(
    transformers=[
        ('onehot', encoder_categoricas, categoricas),
        ('minmax', encoder_numericas, numericas)
    ],
    verbose_feature_names_out=False
).set_output(transform="pandas")
```

#### 3.2 Holdout

Realizamos un holdout 80-20 de manera estratificada para mantener balance de clases:

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

Preprocesamos, entrenamos, predecimos y mostramos métricas generales:

```python
# Clasificador dummy
dummy = DummyClassifier(strategy='stratified',random_state=42)

# Pipeline de clasificación dummy
pipe_dummy = Pipeline([
    ('preprocesamiento', preprocesamiento),
    ('dummy', dummy)
])

# Entrenamiento y predicción
pipe_dummy.fit(X_train, y_train)
y_pred_dummy = pipe_dummy.predict(X_test)

# Impresión de resultados
print('Dummy Classifier:\n')
print(classification_report(y_test, y_pred_dummy))
```

```python
# Clasificador knn
knn = KNeighborsClassifier()

# Pipeline de clasificación knn
pipe_knn = Pipeline([
    ('preprocesamiento', preprocesamiento),
    ('knn', knn)
])

# Entrenamiento y predicción
pipe_knn.fit(X_train, y_train)
y_pred_knn = pipe_knn.predict(X_test)

# Impresión de resultados
print('KNeighbors Classifier:\n')
print(classification_report(y_test, y_pred_knn))
```

```python
# Clasificador dtc
dtc = DecisionTreeClassifier(random_state=42)

# Pipeline de clasificación dtc
pipe_dtc = Pipeline([
    ('preprocesamiento', preprocesamiento),
    ('dtc', dtc)
])

# Entrenamiento y predicción
pipe_dtc.fit(X_train, y_train)
y_pred_dtc = pipe_dtc.predict(X_test)

# Impresión de resultados
print('Decision Tree Classifier:\n')
print(classification_report(y_test, y_pred_dtc))
```

Creamos dos nuevos pipelines para la búsqueda de hiperparámetros: uno para LightGBM Classifier (mejor modelo obtenido) y otro para Random Forest Classifier (segundo mejor modelo obtenido). Estos pipelines serán iguales a los creados anteriormente pero con un paso intermedio de selección de atributos/características (donde el parámetro de cuántas características se seleccionan será parametrizable y configurado por el optimizador de hiperparámetros) mediante el método SelectPercentile de sklearn con función de puntaje f_classif (la cual calcula el puntaje ANOVA F de cada característica en relación con las clases objetivo):

```python
# Librerías
from sklearn.feature_selection import SelectPercentile, f_classif

# Selector de features
selector = SelectPercentile(score_func=f_classif, percentile=80)

# Clasificador LightGBM y Random Forest
lgbmc = LGBMClassifier(random_state=42)
rfc = RandomForestClassifier(random_state=42)

# Pipeline de búsqueda para LightGBM
pipe_lgbmc_grid = Pipeline([
    ('preprocesamiento', preprocesamiento),
    ('seleccion', selector),
    ('lgbmc', lgbmc)
])

# Pipeline de búsqueda para Random Forest
pipe_rfc_grid = Pipeline([
    ('preprocesamiento', preprocesamiento),
    ('seleccion', selector),
    ('rfc', rfc)
])
```

Creamos las grillas de parámetros para cada modelo:

- LightGBM Classifier: 4 posibilidades en Select Percentile (20,40,60,80),

- Random Forest Classifier: 4 posibilidades en Select Percentile (20,40,60,80),


```python
params_lgbmc = [
    {
        'seleccion__percentile': [20,
                                  50,
                                  60,
                                  80],
        'lgbmc__max_depth': [5,
                             10,
                             20],
        'lgbmc__num_leaves': [10,
                              20,
                              30],
        'lgbmc__learning_rate': [0.1,
                                 0.05,
                                 0.01]
    }
]

params_rfc = [
    {
        'seleccion__percentile': [20,
                                  40,
                                  60,
                                  80],
        'rfc__max_depth': [5,
                           10,
                           20],
        'rfc__min_samples_split': [2,
                                   3,
                                   4]
    }
]
```

Importamos librerías:

```python
from sklearn.experimental import enable_halving_search_cv
from sklearn.model_selection import HalvingGridSearchCV
```

Búsqueda de parámetros para lgbmc:

```python
gs_lgbmc = HalvingGridSearchCV(pipe_lgbmc_grid, params_lgbmc, n_jobs=-1, scoring='f1_weighted', verbose=10).fit(X_train, y_train)
```

Búsqueda de parámetros para rfc:

```python
gs_rfc = HalvingGridSearchCV(pipe_rfc_grid, params_rfc, n_jobs=-1, scoring='f1_weighted', verbose=10).fit(X_train, y_train)
```

Mediante la búsqueda de parámetros se obtuvo que el mejor estimador en cada modelo corresponde a:

- LightGBM Classifier: Percentil 40% para la selección de mejores features, 0.05 de learning rate, 20 de profundidad máxima y 10 de número máximo de hojas que puede tener un árbol. Entrega un f1-score weighted average en entrenamiento de 0.759.
