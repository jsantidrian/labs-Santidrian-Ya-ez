1. Git es un DVCS (Sistema de Control de Versiones Distribuido) que registra los cambios en un archivo a lo largo del tiempo utilizando buenas prácticas, como la replicación completa del repositorio realizando una copia completa de los datos, permitiendo establecer varios flujos de trabajo a diferencia de otros sistemas de control de versiones.

2. La primera diferencia radica en su mera definición: Git es un software, mientras que Github es una página web.
La segunda diferencia radica en su modo de uso: Git se utiliza como una herramienta de línea de comando, mientras que Github se utiliza mediante una interfaz gráfica.
La tercera diferencia radica en su función: Git crea un repositorio local para trackear cambios locales de archivos, mientras que Github es un espacio para subir una copia de los repositorios de Git.

3. Cada objeto de Git tiene un identificador único, el cual fue almacenado mediante un método de criptografía, y este es llamado Hash.
