### 2.7 Segmentación Etárea

Para realizar la segmentación etárea, separaremos según las edades de los clientes que se encuentran en la variable **customer_age** (las cuales están redondeadas a la década) de la siguiente forma:

- Joven (Menor a 18 años) -> Todos los que tienen valor 10
- Adulto-Joven (18 - 26 años) -> Todos los que tienen valor 20
- Adulto (27- 59 años) -> Todos los que tienen valor 30, 40 o 50
- Persona Mayor (60 años o mas) -> Todos los que tienen valor 60, 70, 80 o 90.

A continuación pasaremos a hacer esta segmentación etárea y a agregarla como una nueva variable al dataframe:

```python
# Función que realiza la segmentación etárea
def segmentacion(x):
  if x == 10:
    return 'Joven (Menor a 18 años)'
  elif x == 20:
    return 'Adulto-Joven (18-26 años)'
  elif x in (30,40,50):
    return 'Adulto (27-59 años)' 
  elif x in (60,70,80,90):
    return 'Persona Mayor (60 años o más)'

# Creamos nueva columna en el dataframe que captura la segmentación etárea
df_limpio['segmentacion_etarea'] = df_limpio['customer_age'].apply(segmentacion)
```

## 4. Preprocesamiento

### 4.1 Declarar `ColumnTransformer`

Primero escalaremos los datos numéricos utilizando MinMaxScaler(), esto para prevenir sesgos en nuestro modelo y reducir la influencia de outliers.

Luego transformaremos los datos categóricos (tipo object) del dataset (*payment_type, employment_status, housing_status, source, device_os, segmentacion_etarea*) separando en naturaleza nominal y ordinal:

- Nominales -> No tienen un orden inherente (*payment_type, source, device_os*). Se aplicará One Hot Encoder para preprocesarlas.

- Ordinales -> Tienen un orden inherente (*employment_status, housing_status, segmentacion_etarea*). Se aplicará Ordinal Encoder para preprocesarlas.

```python
# Importar librería
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder, OrdinalEncoder

# Preprocesadores para datos numéricos, nominales y ordinales
preprocesador_numerical = MinMaxScaler()
preprocesador_nominal = OneHotEncoder(sparse=False)
preprocesador_ordinal = OrdinalEncoder()

# Columnas numéricas, nominales y ordinales
numerical = df_limpio.select_dtypes(include=['float64']).columns 
nominal = ['payment_type','source','device_os']
ordinal = ['employment_status','housing_status','segmentacion_etarea']

# Columnas no transformadas
pasadas = ['id','fraud_bool']

# ColumnTransformer
preprocesador = ColumnTransformer(transformers=[
    ('preprocesador_numerical', preprocesador_numerical, numerical),
    ('preprocesador_nominal', preprocesador_nominal, nominal),
    ('preprocesador_ordinal', preprocesador_ordinal, ordinal),
    ('passthrough', 'passthrough', pasadas)
])
```

### 4.2 Transformar datos

```python
# Pipeline
pipeline = Pipeline([
    ('preprocesador',preprocesador)
])

# Output es un dataframe
pipeline.set_output(transform='pandas')

# Aplicamos el pipeline a los datos
df_transformed = (
    pipeline.fit_transform(df_limpio)
    .reset_index()
    .drop(columns=['index','preprocesador_numerical__fraud_bool'],axis=1)
)
```

Visualizamos:

```python
df_transformed
```

Se aprecia que todos los datos numéricos originales están escalados mediante *MinMaxScaler* (a excepción de *id* y *fraud_bool*) y que los datos categóricos se transformaron utilizando *OneHotEncoder* (para los nominales) y *OrdinalEncoder* (para los ordinales), por lo que está terminado el preprocesamiento.

### 5.2 Proyectar y agregar proyecciones al dataframe de muestreo

Utilizaremos PCA para encontrar las 2 componentes principales:

```python
# Obtención de las 2 componentes principales mediante PCA
componentes = PCA(n_components=2,random_state=0)

# Quitamos variable id y fraud_bool de df_muestra
df_muestra_2 = df_muestra.drop(columns=['passthrough__id','passthrough__fraud_bool'],axis=1)

# Aplicamos PCA
componentes_principales = componentes.fit_transform(df_muestra_2)

# Creamos un nuevo dataframe con las componentes principales
df_componentes = pd.DataFrame(data=componentes_principales, columns=['Componente 1', 'Componente 2'])

# Visualizamos meidante un scatter plot
fig = px.scatter(df_componentes, x='Componente 1', y='Componente 2')
fig.update_layout(
    title = 'Visualización en baja dimensionalidad de una muestra aleatoria de los datos preprocesados'
)
fig.show()
```
